```python
import pandas as pd
from collections import defaultdict

# Assuming you already have:
# - expr_df: your full expression DataFrame with gene columns (some with '///' separated names)
# - metabolic_set: a Python set containing all metabolic gene symbols for quick membership check

# Step 1: Convert column names to strings (important if some are not strings)
original_cols = final_data.columns.astype(str)

# Helper function: splits '///' separated gene symbols and trims whitespace
def extract_genes(col_name):
    # If col_name is NaN or not a string, return empty list
    if not isinstance(col_name, str):
        return []
    return [g.strip() for g in col_name.split('///')]

# Step 2: Build metadata for each column:
# - original_name: original column name
# - normalized_name: joined gene symbols with underscore (used for final renaming)
# - genes: list of all gene symbols in the column
# - matched_genes: genes from the column that are metabolic (exist in metabolic_set)
# - n_matches: number of matched metabolic genes
col_metadata = []

for col in original_cols:
    genes = extract_genes(col)
    matched = [g for g in genes if g in l3]
    
    col_metadata.append({
        'original_name': col,
        'normalized_name': '_'.join(genes),
        'genes': genes,
        'matched_genes': matched,
        'n_matches': len(matched)
    })

# Step 3: Keep only columns with exactly one metabolic gene match
filtered_meta = [entry for entry in col_metadata if entry['n_matches'] == 1]

# Step 4: Handle duplicate normalized names by appending _1, _2, etc.
name_counter = defaultdict(int)

for entry in filtered_meta:
    base_name = entry['normalized_name']
    name_counter[base_name] += 1
    
    if name_counter[base_name] > 1:
        entry['final_name'] = f"{base_name}_{name_counter[base_name]}"
    else:
        entry['final_name'] = base_name
cols_to_keep = [entry['original_name'] for entry in filtered_meta]

# Check for duplicates in cols_to_keep
print(f"Original cols_to_keep length: {len(cols_to_keep)}")
cols_to_keep_unique = list(dict.fromkeys(cols_to_keep))
print(f"Unique cols_to_keep length: {len(cols_to_keep_unique)}")

# If duplicates removed, filtered_meta needs update too:
if len(cols_to_keep_unique) < len(cols_to_keep):
    print("Duplicates found and removed from cols_to_keep!")
    # Filter filtered_meta accordingly
    filtered_meta = [entry for i, entry in enumerate(filtered_meta) if cols_to_keep[i] in cols_to_keep_unique]
    cols_to_keep = cols_to_keep_unique

filtered_expr_df = final_data.loc[:, cols_to_keep].copy()

print(f"Filtered DataFrame columns: {filtered_expr_df.shape[1]}")
print(f"Length of new column names: {len([entry['final_name'] for entry in filtered_meta])}")

# Now assign final column names
filtered_expr_df.columns = [entry['final_name'] for entry in filtered_meta]

# Step 6: Create metadata DataFrame summarizing final names, original names, and matched metabolic gene
meta_df = pd.DataFrame(filtered_meta)[['final_name', 'original_name', 'matched_genes']].copy()
# Since each column has exactly one matched gene, extract that gene into a separate column
meta_df['matched_gene'] = meta_df['matched_genes'].apply(lambda x: x[0] if x else None)
meta_df.drop(columns=['matched_genes'], inplace=True)

# Now you have:
# - filtered_expr_df: expression data only for columns with exactly one metabolic gene, columns renamed uniquely
# - meta_df: metadata table linking final column names to original columns and matched metabolic gene
```
