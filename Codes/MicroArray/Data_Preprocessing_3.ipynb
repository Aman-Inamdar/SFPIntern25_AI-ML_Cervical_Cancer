```python
import pandas as pd
```

```python
df=pd.read_csv("Merged_Datasets_Combat.csv")
df.head()
```

```python
l1=list(df.columns)
l1[0]="ID"
```

```python
l1=[x.upper().replace(".CEL","") for x in l1]
```

```python
df.columns=l1
```

```python
df.head()
```

```python

df=df.set_index("ID")
```

```python
df=df.T
```

```python
df.head()
```

```python
meta_final=pd.read_csv("Metadata_2.csv")
meta_final.head()
```

```python
#Missing values in Age: 5 (Filling with mean of the all values)
meta_final["Age"]=meta_final["Age"].fillna(meta_final["Age"].mean())
```

```python
#selecting required metadata columns from the metadata
meta=meta_final[["ID","Age","Cancer","HPV_Status"]]
```

```python
meta=meta.set_index("ID")
```

```python
meta.head()
```

```python
#Checking value counts in the cancer column
meta["Cancer"].value_counts()
#Implies that: Cancerous->53 and Non-Cancerous->32
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.countplot(x=meta["Cancer"],data=meta)
```

```python
#checking the value counts in HPV_Status column
meta["HPV_Status"].value_counts()
```

```python
sns.countplot(x=meta["HPV_Status"],data=meta)
```

```python
df_1=pd.get_dummies(meta["HPV_Status"],drop_first=True,dtype=int)
```

```python
META=pd.merge(meta.drop("HPV_Status",axis=1),df_1,left_index=True,right_index=True)
```

```python
META.head()
```

```python
df_final=pd.merge(df,META,left_index=True,right_index=True)
```
