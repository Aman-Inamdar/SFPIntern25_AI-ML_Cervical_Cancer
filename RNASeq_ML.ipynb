```python
import pandas as pd
import numpy as np 
```

Data Preprocessing Normal and Cancerous Samples


```python
#loading TCGA-GDC dataset no. of samples : 310 (all cancerous)
df_cancer=pd.read_csv("TCGA-CESC.star_tpm.tsv/TCGA-CESC.star_tpm.tsv",sep="\t")
df_cancer.head(2)
```

```python
df_cancer=df_cancer.set_index("Ensembl_ID")
```

```python
df_cancer.head(1)
```

```python
# Mapping to genes
df_genes=pd.read_csv("TCGA-CESC.star_tpm.tsv/gencode.v36.annotation.gtf.gene.probemap",sep="\t")
df_genes.head(1)
```

```python
#selecting only id and gene columns
df_gene_selected=df_genes[["id","gene"]]
df_gene_selected=df_gene_selected.set_index("id")
```

```python
df_gene_selected.head(1)
```

```python
#mapping with the csv file 
df_annotated=pd.merge(df_cancer,df_gene_selected,left_index=True,right_index=True)
```

```python
df_annotated.head(1)
```

```python
df_annotated=df_annotated.reset_index()
```

```python
df_with_genes=df_annotated.drop("index",axis=1)
```

```python
df_with_genes.head(1)
```

```python
df_with_genes=df_with_genes.set_index("gene")
```

```python
df_with_genes=df_with_genes.T
df_with_genes.head(1)
```

```python
df_with_genes=df_with_genes.reset_index()
```

```python
df_with_genes.head(1)
```

```python
df_with_genes["Indicator_1"]=df_with_genes["index"].str.split("-").str[3]
```

```python
df_with_genes.head(1)
```

```python
df_with_genes["Label"]=df_with_genes["Indicator_1"].str.split("").str[2]
```

```python
df_with_genes.head(1)
```

```python
df_with_genes["Label"]=df_with_genes["Label"].astype(int)
```

```python
df_with_genes.drop(["index","Indicator_1"],axis=1,inplace=True)
```

```python
df_with_genes.head(1)
```

```python
df_with_genes["Label"].value_counts()
```

```python
# 6 -> represents the metastasis and 1 -> represents the primary solid tumor (cancer)
df_with_genes["Label"]=df_with_genes["Label"].replace(6,1)
```

```python
df_with_genes.to_csv("Dataset_with_labels.csv")
```

```python
df_with_genes.drop("Label",axis=1).to_csv("Dataset_without_Labels.csv")
```

```python
l1=(df_1["GeneID"]).tolist() #l1 represents the entrez ids 
```

```python
import mygene
# Query MyGene.info
mg = mygene.MyGeneInfo()
results = mg.querymany(l1, scopes='entrezgene', fields='symbol', species='human')

# Extract and format results
data = [{'EntrezID': r['query'], 'GeneSymbol': r.get('symbol', 'NA')} for r in results]

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('entrez_to_symbol.csv', index=False)
```

```python
df["EntrezID"]=df["EntrezID"].astype(int)
```

```python
#1691 have no gene names associated for the respective entrez id 
df_1_merged=pd.merge(df_1,df,left_on="GeneID",right_on="EntrezID")
df_1_merged.head()
```
